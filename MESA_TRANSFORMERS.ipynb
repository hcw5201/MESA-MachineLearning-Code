```python
#Code checked: 09/18/2024, Dumi

# Import libraries
import numpy as np
from tensorflow.keras import layers, models, optimizers, losses
from sklearn.metrics import r2_score
import tensorflow.keras.backend as K

import matplotlib.pyplot as plt
import time

import tensorflow as tf
from tensorflow.keras.layers import Dense, Input, LayerNormalization, MultiHeadAttention, Dropout, Reshape
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Load and preprocess
df = pd.read_csv('new_vega_rotation_overshooting_alpha_mlt_test_no_num_BIGGER_num_gt_180.csv')

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[['mass', 'z','fov','mlt', 'age', 'teff', 'lum']], df['log_k'], test_size=0.2, random_state=123)

# Define the Transformer Model for Regression
def transformer_block(inputs, head_size, num_heads, ff_dim, dropout=0):
    # Multi-Head Attention Layer
    attention_output = MultiHeadAttention(num_heads=num_heads, key_dim=head_size)(inputs, inputs)
    attention_output = Dropout(dropout)(attention_output)
    attention_output = LayerNormalization(epsilon=1e-6)(attention_output)

    # Feed Forward Layer
    ff_output = Dense(ff_dim, activation="relu")(attention_output)
    ff_output = Dropout(dropout)(ff_output)
    ff_output = LayerNormalization(epsilon=1e-6)(ff_output)

    return ff_output

# Model architecture
def build_model(input_shape, head_size, num_heads, ff_dim, num_transformer_blocks, mlp_units, dropout=0, mlp_dropout=0):
    inputs = Input(shape=input_shape)

    # Reshape input to 3D tensor (batch_size, sequence_length=1, num_features)
    x = Reshape((1, input_shape[0]))(inputs)

    for _ in range(num_transformer_blocks):
        x = transformer_block(x, head_size, num_heads, ff_dim, dropout)

    # Flatten after transformer blocks
    x = tf.keras.layers.Flatten()(x)

    x = Dense(mlp_units[0], activation="relu")(x)
    x = Dropout(mlp_dropout)(x)
    x = Dense(mlp_units[1], activation="relu")(x)
    x = Dropout(mlp_dropout)(x)

    # Output layer for regression
    outputs = Dense(1, activation="linear")(x)

    return tf.keras.Model(inputs, outputs)

# Set hyperparameters
input_shape = (X_train.shape[1],)  # Number of features
model = build_model(input_shape, head_size=64, num_heads=2, ff_dim=128, num_transformer_blocks=2,
                    mlp_units=[128, 64], dropout=0.1, mlp_dropout=0.1)

# List of optimizers to try
optimizers_to_try = [
    tf.keras.optimizers.Adam(learning_rate=0.001),
    tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9),
    tf.keras.optimizers.Adagrad(learning_rate=0.01),
    tf.keras.optimizers.RMSprop(learning_rate=0.001),
    tf.keras.optimizers.Adadelta(learning_rate=1.0),
    tf.keras.optimizers.Nadam(learning_rate=0.002)
]


# Define R-squared as a custom metric
def r2(y_true, y_pred):
    SS_res = K.sum(K.square(y_true - y_pred))
    SS_tot = K.sum(K.square(y_true - K.mean(y_true)))
    return 1 - SS_res / (SS_tot + K.epsilon())

# Rename lists to avoid conflicts
val_losses = []
val_mses = []
val_maes = []
val_r2s = []

# Train with each optimizer
for i, optimizer_selected in enumerate(optimizers_to_try):
    # Build and compile the model
    model = build_model(input_shape, head_size=64, num_heads=2, ff_dim=128, num_transformer_blocks=2,
                        mlp_units=[128, 64], dropout=0.1, mlp_dropout=0.1)
    model.compile(optimizer=optimizer_selected, loss="mse", metrics=["mse", "mae", r2])

    # Train the model
    history = model.fit(X_train, y_train, validation_split=0.2, epochs=20, batch_size=64)

    # Collect metrics for plotting and comparison
    val_losses.append(history.history['val_loss'])
    val_mses.append(history.history['val_mse'])
    val_maes.append(history.history['val_mae'])
    val_r2s.append(history.history['val_r2'])

# Plot epoch vs validation loss for comparison
x = range(1, 21)
plt.figure()
for j, optimizer in enumerate(['Adam', 'SGD', 'Adagrad', 'RMSprop', 'Adadelta', 'Nadam']):
    plt.plot(x, val_losses[j], label=optimizer)
plt.xlabel('Epoch')
plt.ylabel('Validation Loss')
plt.title('Epoch vs Validation Loss for Transformer Model')
plt.legend()
plt.show()

# Print results for each optimizer
for j, optimizer in enumerate(['Adam', 'SGD', 'Adagrad', 'RMSprop', 'Adadelta', 'Nadam']):
    print(f"Results with optimizer [{optimizer}] MSE: {val_mses[j][-1]} MAE: {val_maes[j][-1]} R2: {val_r2s[j][-1]})")


# # Compute R^2 score
# r2 = r2_score(y_test.numpy(), y_test)
# print(f'R^2 Score on test set: {r2}')
```

```python
# Testing the Performance of our Model
from sklearn.metrics import mean_squared_error, r2_score

# import time
# Predict the values using the model
start_time = time.time()
y_pred = model.predict(X_test)
end_time = time.time()
test_time = end_time - start_time
print(f"Testing time: {test_time:.2f} seconds")


# Calculate MSE
mse = mean_squared_error(y_test, y_pred)

# Calculate R²
r2 = r2_score(y_test, y_pred)

# Print results
print(f"Test MSE: {mse}")
print(f"Test R²: {r2}")

# Plot actual vs predicted values
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
plt.title('Actual vs Predicted Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')

plt.show()
```

## Feature Importance

```python
# Define a list of feature names for reference and labeling
features = ['mass', 'z', 'fov', 'mlt', 'age', 'teff', 'lum']

# Function to calculate permutation feature importance
def permutation_feature_importance(model, X_test, y_test, n_repeats=10):
    baseline_performance = model.evaluate(X_test, y_test, verbose=0)[0]
    
    importances = np.zeros(X_test.shape[1])
    
    # Loop through each feature in the test set
    for i in range(X_test.shape[1]):
        scores = []  
        X_test_permuted = X_test.to_numpy().copy() 
        
        for _ in range(n_repeats):
            np.random.shuffle(X_test_permuted[:, i])
            
            permuted_performance = model.evaluate(X_test_permuted, y_test, verbose=0)[0]
            
            scores.append(permuted_performance - baseline_performance)
        
        importances[i] = np.mean(scores)
    
    return importances  
    
# Compute the feature importances
feature_importances = permutation_feature_importance(model, X_test, y_test)

# Print the feature importance scores
for feature, importance in zip(features, feature_importances):
    print(f"Feature: {feature}, Importance Score: {importance:.4f}")

# Plot the feature importances as a bar chart
plt.bar(features, feature_importances)
plt.xlabel('Features')  
plt.ylabel('Importance Score')  
plt.title('Feature Importance for Transformers')  
plt.xticks(rotation=45) 
plt.show()  
```
